```python
a= input()
list1= [i for i in a[1:-1].split(",")]
dict1={}
for i in list1:
    b= i.split(":")
    keys=b[0]
    values=int(b[1])
    dict1[keys]=values
    s=0
    count=0
for k,v in dict1.items():
    s+=v
    count+=1
avg= s//count
print(avg)
```
